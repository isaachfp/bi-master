#Vamos a ver DBSCAN

##Primeiro as bibliotecas

```
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

##vamos criar dados de maneira probabilistica

```
#Determine centroids
centers = [[0.5, 2], [-1, -1], [1.5, -1]]
#Create dataset
# Cria 400 exemplos, de maneira rand√¥mica, com os centros indicados
X, y = make_blobs(n_samples=400, centers=centers, 
                  cluster_std=0.5, random_state=0)
# Assim geramos dados a partir de centroids
#Normalize the values
X = StandardScaler().fit_transform(X)
```

Agora vemos um grafico

```
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='Paired')
```

```
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.4, min_samples=10)
db.fit(X)
#Os ajustes de eps e min_samples √© que v√£o determinar a quantidade de outliers
#Testei com 0.4 e 0.5 e mantendo o n√∫mero de samples, foi o que gerou menos outliers
```

## N√£o precisamos especificar o n√∫mero de clusters para DBSCAN, o que √© uma grande vantagem do DBSCAN sobre o cluster k-means.

Vamos visualizar os clusters determinados pelo DBSCAN:

```
y_pred = db.fit_predict(X)
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y_pred, cmap='Paired')
plt.title("Clusters determined by DBSCAN")
```
